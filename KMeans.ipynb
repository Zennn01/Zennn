```python
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
```

```python
df = pd.read_csv('Mall_Customers.csv')
```

```python
X = df[['Annual Income (k$)', 'Spending Score (1-100)']]
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python

kmeans = KMeans(n_clusters=5, init='k-means++', random_state=42)
df['Cluster'] = kmeans.fit_predict(X_scaled)
```

```python
centroids = scaler.inverse_transform(kmeans.cluster_centers_)
```

```python
cluster_labels = []
for income, score in centroids:
    if income > 70 and score > 60:
        cluster_labels.append("üí∞ Penghasilan & belanja tinggi")
    elif income > 70 and score < 40:
        cluster_labels.append("‚ùì Penghasilan tinggi, belanja rendah")
    elif income < 40 and score > 60:
        cluster_labels.append("üì£ Penghasilan rendah, belanja tinggi")
    elif income < 40 and score < 40:
        cluster_labels.append("üßç Penghasilan & belanja rendah")
    else:
        cluster_labels.append("‚öñÔ∏è Rata-rata")
```

```python
label_mapping = dict(zip(range(5), cluster_labels))
df['Deskripsi Cluster'] = df['Cluster'].map(label_mapping)
```

```python
hasil = df[['CustomerID', 'Gender', 'Age', 'Annual Income (k$)', 'Spending Score (1-100)', 'Cluster', 'Deskripsi Cluster']]
hasil.groupby('Deskripsi Cluster').head(10).sort_values('Cluster').reset_index(drop=True)
```

```python
df.to_csv('Hasil_KMeans_Mall_Customers.csv', index=False)
```
